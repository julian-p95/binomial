```python
def binom(n,p):
    X=0
    for k in range(n):
        X = X + floor(random() + p)
    return X
```

```python
carnets_25 = [binom(50,p) for k in range(25)]
```

```python
def comptage_carnets(Liste_carnets):
    n = 0
    for a in Liste_carnets:
        if a>=2:
            n = n + 1
    return n
```

```python
comptage_carnets(carnets_25)
```

```python
L = []
for i in range(10000):
    carnets_25 = [binom(50,p) for k in range(25)]
    nb = comptage_carnets(carnets_25)
    L.append(nb)
    
F = [L.count(k) / 10000 for k in range(max(L) + 1)]
print(F)
```

```python
import pylab as pl
pl.style.use('bmh')

pl.bar(range(max(L) + 1), F, width=0.5, color='black', label='Simulations')
pl.xticks(range(max(L) + 1))
pl.xlabel('Valeurs de SNS', fontsize=14)
pl.show()
```

```python
import scipy.stats as sc
import pylab as pl

p = 40529 / 150000
N = sc.binom(25, p)
P = [N.pmf(k) for k in range(max(L) + 1)]

pl.bar(range(max(L) + 1), F, width=0.5, color='black', label='Simulations')
pl.bar(range(max(L) + 1), P, alpha=0.5, color='red', label='Theory')
pl.xticks(range(max(L) + 1))
pl.xlabel('Values of SNS', fontsize=14)
pl.legend(fontsize=13)
pl.show()
```
